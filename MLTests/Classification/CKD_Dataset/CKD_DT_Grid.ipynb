```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rootPath = "../../../../../"
dataset=pd.read_csv(rootPath+"Datasets/CKD.csv")
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
dataset["classification_yes"].value_counts()
```

```python
independent=dataset[
    ['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv','wc',
     'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal','pcc_present',
       'ba_present', 'htn_yes', 'dm_yes', 'cad_yes','appet_yes','pe_yes', 'ane_yes']
]
dependent=dataset["classification_yes"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
param_grid={'criterion':['poisson','squared_error','gini','entropy'],'max_features':['auto','sqrt','log2']}
grid=GridSearchCV(DecisionTreeClassifier(),param_grid,refit=True,verbose=3,n_jobs=1)
grid.fit(X_train,y_train)
```

```python
y_pred=grid.predict(X_test)
y_pred
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(X_test)[:,1])
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,y_pred)
```

```python
print(clf_report)
```
