```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rootPath = "../../../../../"
dataset=pd.read_csv(rootPath+"Datasets/CKD.csv")
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
dataset["classification_yes"].value_counts()
```

```python
independent=dataset[
    ['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv','wc',
     'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal','pcc_present',
       'ba_present', 'htn_yes', 'dm_yes', 'cad_yes','appet_yes','pe_yes', 'ane_yes']
]
dependent=dataset["classification_yes"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
param_grid={'kernel': ["linear", "rbf"],'C': [0.1,1]}#"poly","sigmoid"],'gamma':['auto','scale'], 'C':[10,100] ##including this makes this perform super slow. so excluded it via comments.
grid=GridSearchCV(SVC(probability=True,random_state=7,cache_size=1000),param_grid,scoring='r2',cv=5,refit=True,verbose=3,n_jobs=1)#,scoring='f1_weighted'
grid.fit(X_train,y_train)
```

```python
print("Best parameters found:",grid.best_params_)
print("Best cross-validation score:",grid.best_score_)
```

```python
#results_df = pd.DataFrame(grid.cv_results_)
#print(results_df)
#print(results_df[['params', 'mean_test_score', 'std_test_score', 'rank_test_score']])
table=pd.DataFrame.from_dict(grid.cv_results_)
table
```

```python
y_pred=grid.predict(X_test)
y_pred
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(X_test)[:,1])
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,y_pred)
```
